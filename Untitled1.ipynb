```python
import numpy as np
import matplotlib.pyplot as plt

def mandelbrot(c, max_iter):
    z = 0
    for n in range(max_iter):
        if abs(z) > 2:
            return n
        z = z*z + c
    return 0

def mandelbrot_set(xmin, xmax, ymin, ymax, width, height, max_iter):
    x = np.linspace(xmin, xmax, width)
    y = np.linspace(ymin, ymax, height)
    C = np.meshgrid(x, y)
    C = np.array(C[0] + 1j * C[1]).flatten()

    mandelbrot_vector = np.vectorize(mandelbrot)
    result = mandelbrot_vector(C, max_iter)
    result = result.reshape((height, width))

    return result

def plot_mandelbrot(xmin, xmax, ymin, ymax, width=10, height=10, max_iter=256):
    result = mandelbrot_set(xmin, xmax, ymin, ymax, width*100, height*100, max_iter)
    plt.imshow(result, extent=(xmin, xmax, ymin, ymax), cmap='jet')
    plt.colorbar()
    plt.title("Mandelbrot Set")
    plt.show()

plot_mandelbrot(-2.0, 0.5, -1.25, 1.25)
```

```python
def julia(c, z, max_iter):
    for n in range(max_iter):
        if abs(z) > 2:
            return n
        z = z*z + c
    return 0

def julia_set(c, xmin, xmax, ymin, ymax, width, height, max_iter):
    x = np.linspace(xmin, xmax, width)
    y = np.linspace(ymin, ymax, height)
    Z = np.meshgrid(x, y)
    Z = np.array(Z[0] + 1j * Z[1]).flatten() # flatten x and y to single vector?

    julia_vector = np.vectorize(julia)
    result = julia_vector(c, Z, max_iter)
    result = result.reshape((height, width)) #

    return result

def plot_julia(c, xmin, xmax, ymin, ymax, width=10, height=10, max_iter=256):
    result = julia_set(c, xmin, xmax, ymin, ymax, width*100, height*100, max_iter)
    plt.imshow(result, extent=(xmin, xmax, ymin, ymax), cmap='jet')
    plt.colorbar()
    plt.title("Julia Set (c = {})".format(c))
    plt.show()

# Example usage:
plot_julia(-0.8 + 0.156j, -1.5, 1.5, -1.5, 1.5)
```
